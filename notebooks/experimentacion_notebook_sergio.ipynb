```python
def get_stemmer(text):
  palabras = [palabra.lower() for palabra in re.findall(r'\w+', text.lower())]
  text_lista = []
  for palabra in palabras:
    nueva = lancStemmer.stem(palabra)
    text_lista.append(nueva)
  nuevo_texto = ' '.join(text_lista)
  return nuevo_texto
```

```python
from nltk.tokenize import word_tokenize
```

```python
def pre_procesamiento(parrafo):
    # Eliminar caracteres no alfanuméricos
    parrafo = re.sub(r'[^\w\s]', '', parrafo)
    # Tokenizar el texto
    tokenized_text = word_tokenize(parrafo)
    # Inicializar el stemmer de Lancaster
    stemmer = LancasterStemmer()
    # Realizar stemming en cada palabra
    stemmed_text = [stemmer.stem(word) for word in tokenized_text]
    # Retornar el texto preprocesado
    return stemmed_text
```

```python
def get_grams(text, n):
  text = get_stemmer(text) #pre-procesa el parrafo
  text = re.findall(r"\w+", text) #separa los caracteres pre-procesados del parrafo en listas
  grams = ngrams(text,n) #genera los ngrams
  result = []
  for ng in grams:
    result.append(' '.join(ng)) #agrega los ngrams en una lista llamada result
  return result
```

```python
#Read files
import os

def read_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            with open(os.path.join(directory, filename), 'r') as f:
                content = f.read()
                texto_stems = [content]
                texto_stems.append(texto_stems)
    return texto_stems
```

```python
def matriz_parrafos(gramas1, gramas2):
  grams_palabras = set(gramas1 + gramas2) #set de palabras de ambos ngrams
  grams_juntos = [gramas1, gramas2] #lista con ambas listas de los ngrams de cada parrafo
  matriz = []
  for grama in grams_juntos:
    vector = []
    for palabra in grams_palabras:
      vector.append(1 if palabra in grama else 0) #compara las palabras de los grams a la palabra y agrega o un 1 o un 0 al vector del parrafo
    matriz.append(vector)
  return matriz
```

```python
def similitud_documento(doc1, doc2):
    unigrams_doc1 = get_grams(doc1, 1)
    bigrams_doc1 = get_grams(doc1, 2)
    trigrams_doc1 = get_grams(doc1, 3)

    unigrams_doc2 = get_grams(doc2, 1)
    bigrams_doc2 = get_grams(doc2, 2)
    trigrams_doc2 = get_grams(doc2, 3)

    similitud_unigrams = cosine_similarity(matriz_parrafos(unigrams_doc1, unigrams_doc2))
    similitud_bigrams = cosine_similarity(matriz_parrafos(bigrams_doc1, bigrams_doc2))
    similitud_trigrams = cosine_similarity(matriz_parrafos(trigrams_doc1, trigrams_doc2))

    print("Similitud de coseno (unigrams) entre doc1 y doc2:", similitud_unigrams[0][1])
    print("Similitud de coseno (bigrams) entre doc1 y doc2:", similitud_bigrams[0][1])
    print("Similitud de coseno (trigrams) entre doc1 y doc2:", similitud_trigrams[0][1])
```

```python
def get_grams_mod(text, n):
    text = re.findall(r"\w+", text.lower()) # Convertir texto a minúsculas y separarlo en palabras
    grams = ngrams(text, n) # Generar n-gramas
    result = []
    for ng in grams:
        result.append(' '.join(ng)) # Convertir n-gramas a strings
    return result
```

```python
def similarity_mod(word1, word2):
    if word1 in model.wv.vocab and word2 in model.wv.vocab:
        return model.wv.similarity(word1, word2)
    else:
        return 0
```

```python
def similitud_documento(doc1, doc2):
    unigrams_doc1 = get_grams_mod(doc1, 1)
    bigrams_doc1 = get_grams_mod(doc1, 2)
    trigrams_doc1 = get_grams_mod(doc1, 3)

    unigrams_doc2 = get_grams_mod(doc2, 1)
    bigrams_doc2 = get_grams_mod(doc2, 2)
    trigrams_doc2 = get_grams_mod(doc2, 3)

    # Calcular la similitud usando los embeddings de Word2Vec
    similitud_unigrams = np.mean([similarity_mod(w1, w2) for w1 in unigrams_doc1 for w2 in unigrams_doc2])
    similitud_bigrams = np.mean([similarity_mod(w1, w2) for w1 in bigrams_doc1 for w2 in bigrams_doc2])
    similitud_trigrams = np.mean([similarity_mod(w1, w2) for w1 in trigrams_doc1 for w2 in trigrams_doc2])

    print("Similitud (Word2Vec) entre doc1 y doc2 (unigrams):", similitud_unigrams)
    print("Similitud (Word2Vec) entre doc1 y doc2 (bigrams):", similitud_bigrams)
    print("Similitud (Word2Vec) entre doc1 y doc2 (trigrams):", similitud_trigrams)
```
